```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
def marg(fname):
    df = pd.read_csv(fname)
    df = df.groupby('diff_coef',as_index=False).sum()
    return df['diff_coef'].to_numpy(), df['mean_posterior_occupation'].to_numpy()
```

```python
D,dr = marg('v_pocsv.csv')
_,papa = marg('g_pocsv.csv')
```

```python
fbound_thresh = 0.1

plt.figure(figsize=(6,4))
plt.semilogx(D,dr,color='darkviolet',label='DR')
plt.semilogx(D,papa,color='green',label='PAPA')
plt.tick_params(axis='x', labelsize=16) 
plt.tick_params(axis='y', labelsize=16) 
plt.xlabel('Diffusion coefficient (µm$^2$/s)',fontsize=18)
plt.ylabel('Fraction of molecules',fontsize=18);
plt.legend()

print('DR fbound')
print(dr[D<fbound_thresh].sum())
print('PAPA fbound')
print(papa[D<fbound_thresh].sum())

plt.savefig('papa_vs_dr.png',dpi=300,bbox_inches='tight')
plt.savefig('papa_vs_dr.pdf',bbox_inches='tight')
```
